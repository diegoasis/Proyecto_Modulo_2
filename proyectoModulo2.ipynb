```python
df1 = df1.drop(columns=["keywords", "overview", "tagline"])
df2 = df2.rename(columns={"cast": "cast_members"})
```

```python
def cargar_csv_en_tabla(df, tabla, conn, batch=1000):
    """Inserta un DataFrame en una tabla de TiDB en lotes de `batch` filas."""
    cursor = conn.cursor()
    cols = ", ".join(df.columns)
    placeholders = ", ".join(["%s"] * len(df.columns))
    sql = f"INSERT INTO {tabla} ({cols}) VALUES ({placeholders})"
 
    filas = [tuple(None if pd.isna(v) else v for v in row)
             for row in df.itertuples(index=False, name=None)]
 
    for i in range(0, len(filas), batch):
        cursor.executemany(sql, filas[i:i+batch])
        conn.commit()
        print(f"  insertadas {min(i+batch, len(filas))}/{len(filas)}")
    cursor.close()
 
cargar_csv_en_tabla(df1, "movies", conn, batch=1000)
cargar_csv_en_tabla(df2, "credits", conn, batch=1000)
```

```python
conn = mysql.connector.connect(**mysql_config)

query = """
SELECT
    m.id,
    m.title,
    m.budget,
    m.original_language,
    m.popularity,
    m.release_date,
    m.revenue,
    m.runtime,
    m.status,
    m.vote_average,
    m.vote_count,
    m.genres,
    m.production_companies,
    m.production_countries,
    m.spoken_languages,
    c.cast_members,
    c.crew
FROM proyecto_modulo2.movies m
JOIN proyecto_modulo2.credits c
    ON m.id = c.movie_id
WHERE m.status = 'Released'
"""

df_analisis = pd.read_sql(query, conn)

conn.close()

df_analisis.head()
```
